```python
import torch
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#creating x,y varables for our basic liner regression model
x=torch.linspace(0,10,100).reshape(-1,1)
y=2*x+torch.randn(x.size())
```

```python
x_train,x_test,y_train,y_test= train_test_split(x.numpy(),y.numpy(),test_size=0.2)
```

```python
x_train= torch.tensor(x_train,dtype=torch.float32)
x_test= torch.tensor(x_test,dtype=torch.float32)
y_train= torch.tensor(y_train,dtype=torch.float32)
y_test= torch.tensor(y_test,dtype=torch.float32)
```

```python
#checking if our data is in liner trend or not
plt.scatter(x_train.numpy(),y_train.numpy(),label='train')
plt.scatter(x_test.numpy(),y_test.numpy(),label='test')
plt.legend()
```

```python
x_train= torch.tensor(x_train,dtype=torch.float32)
x_test= torch.tensor(x_test,dtype=torch.float32)
y_train= torch.tensor(y_train,dtype=torch.float32)
y_test= torch.tensor(y_test,dtype=torch.float32)
```

```python
import torch.nn as nn
```

```python
class lrmodel(nn.Module):
  def __init__(self,input_dim,output_dim):
    super(lrmodel,self).__init__()
    self.linear =nn.Linear(input_dim,output_dim)
  def forward(self,x):
    out=self.linear(x)
    return out
#this class contains model initialization and forward pass
```

```python
input_dim=1
output_dim=1
model=lrmodel(input_dim,output_dim)
# here we are makeing model our lrmodel instance to utilize its properties
```

```python
#lets just make predictions based based on just forwasd pass only
with torch.no_grad():
  predicted=model(x_test)#with the use of torch.nograd we are running our model

plt.scatter(x_test.numpy(),y_test.numpy(),label='actual')
plt.scatter(x_test.numpy(),predicted.numpy(),label='predicted')
plt.legend()
```

```python
criterion = nn.MSELoss()
loss=criterion(predicted,y_test)
```

```python
loss
```

```python
import torch.optim as optim
```

```python
optimizer=optim.SGD(model.parameters(),lr=0.01)
```

```python
epoch=100
loss_values=[]

for i in range(epoch):
  model.train() #this indicateds that our model is in training setting
  optimizer.zero_grad()
  output=model(x_train)
  loss=criterion(output,y_train)
  loss.backward()
  optimizer.step()
  loss_values.append(loss.item())
  if(i+1)%10==0:
    print(f'epoch:{epoch+1},loss={loss.item():.4f}')
```

```python
plt.plot(range(epoch),loss_values)
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('loss vs epoch')
plt.show()
```

```python
model.eval()
with torch.no_grad():
  pred_new=model(x_test)
```

```python
# Plot the test data and predictions
plt.scatter(x_test.numpy(), y_test.numpy(), label='Test Data', color='red')
plt.plot(x_test.numpy(), pred_new.numpy(), label='Model Prediction', color='blue')

plt.title('Test Data vs. Model Prediction')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
